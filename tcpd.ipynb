```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt 
data = pd.read_csv('TCPD_GE_Kerala.csv')
df = pd.DataFrame(data)
df.head()
```

```python
data.Sex.value_counts()
```

```python
data["Sex_"] = data.Sex.replace({0: "M", 1: "F", 2:"O",3:"mal"})
x = data.Sex_.value_counts()
(x/x.sum())*100
```

```python
data[data['Year'] == 2004].Sex.value_counts().plot(kind='pie', title='Candidate Gender 2019',figsize=(2, 2), legend=True, labeldistance=None, ylabel='', autopct='%1.1f%%')
```

```python
a = data[data['Year'] == 2004].Sex.value_counts()
b = data[data['Year'] == 2009].Sex.value_counts()
c = data[data['Year'] == 2014].Sex.value_counts()
d = data[data['Year'] == 2019].Sex.value_counts()
fig, axes = plt.subplots(nrows=2, ncols=2)
a.plot(ax=axes[0,0],kind='pie', title='Candidate Gender 2004', figsize=(5, 5),legend=True, labeldistance=None, ylabel='', autopct='%1.1f%%')
b.plot(ax=axes[0,1],kind='pie', title='Candidate Gender 2009',figsize=(5, 5),legend=True, labeldistance=None, ylabel='', autopct='%1.1f%%')
c.plot(ax=axes[1,0],kind='pie', title='Candidate Gender 2014', legend=True, figsize=(5, 5),labeldistance=None, ylabel='', autopct='%1.1f%%')
d.plot(ax=axes[1,1],kind='pie', title='Candidate Gender 2019', legend=True, figsize=(5, 5),labeldistance=None, ylabel='', autopct='%1.1f%%')
```

```python
p = data[data['Year'] == 2004][data['Position'] == 1].Party.value_counts()
q = data[data['Year'] == 2009][data['Position'] == 1].Party.value_counts()
r = data[data['Year'] == 2014][data['Position'] == 1].Party.value_counts()
s = data[data['Year'] == 2019][data['Position'] == 1].Party.value_counts()
fig, axes = plt.subplots(nrows=2, ncols=2)
p.plot(ax=axes[0,0],kind='pie', title='Party candidates who won in 2004', figsize=(10,10),legend=True, labeldistance=None, ylabel='', autopct='%1.1f%%')
q.plot(ax=axes[0,1],kind='pie', title='Party candidates who won in 2009',figsize=(10,10),legend=True, labeldistance=None, ylabel='', autopct='%1.1f%%')
r.plot(ax=axes[1,0],kind='pie', title='Party candidates who won in 2014',figsize=(10,10), legend=True, labeldistance=None, ylabel='', autopct='%1.1f%%')
s.plot(ax=axes[1,1],kind='pie', title='Party candidates who won in 2019', figsize=(10,10),legend=True, labeldistance=None, ylabel='', autopct='%1.1f%%')
```
